# Latino Incarceration Rates in the US, 2000 - 2012

This visualization shows the trend in male and female incarceration rates during the period from 2000 - 2012. The male rate ha declined significantly during the period in question. The female rate increased until around 2007, before experiencing a sharp drop during the rest of the period.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

test_file_h = pd.read_csv('C:\\Users\\fergu\\ImprisonRatesFinal.csv')
```

```python
test_file_h.columns = ['characteristic', 'sex', 'race', 'year', 'rate_per_100k', 'count']
test_file_hx = test_file_h[test_file_h['characteristic'] == 'Hispanic']
#file3 = test_file2.groupby(['characteristic']).mean()
lat_file = test_file_hx[['sex', 'year', 'rate_per_100k']]
lat_male = lat_file[lat_file['sex'] == 'Male']
lat_female = lat_file[lat_file['sex'] == 'Female']
lat_male['testcol'] = lat_male['year'] * 2
print(lat_male)
print(lat_female)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
x = lat_male['year']
y = lat_male['rate_per_100k']
plt.plot(x, y, color='green')
plt.xlabel('Year')
plt.ylabel('Rate per 100,000')
plt.title('Incarceration Rates - Latino Males 2000-2012')

plt.subplot(1,2,2)
i = lat_female['year']
j = lat_female['rate_per_100k']
plt.plot(i, j, color='orange')
plt.xlabel('Year')
plt.ylabel('Rate per 100,000')
plt.title('Incarceration Rates - Latino Females 2000-2012')

plt.show()
```
