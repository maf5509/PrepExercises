# Leading Causes of Death in Texas, 2010 - 2016

The visualization shows which ailments and occurrences led to the most deaths in the state of Texas between 2010 and 2016. Cancer and diseases of the heart are by far the biggest killers, claiming over half of lives lost between them.Interestingly, homicide and suicide only account for a combined 3 % of deaths between them.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

drates = pd.read_csv('C:\\Users\\fergu\\NCHS_-_Leading_Causes_of_Death__United_States.csv')
drates2 = drates[['Year', 'Cause Name', 'State', 'Deaths']]
drates2010_2016 = drates2[drates2['Year'] >= 2010]
dratesTexas = drates2010_2016[drates2010_2016['State'] == 'Texas']
drates_agg = dratesTexas.groupby(['Cause Name']).sum().reset_index()
final = drates_agg[1:]
final.columns = ['cause', 'year', 'deaths']
print(final)

#ax1 = plt.subplots()

pie(final['deaths'], labels=final['cause'], autopct='%1.0f%%', shadow=True, startangle=90)
axis('equal')
plt.show()
```
